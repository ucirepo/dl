```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
training_set=train_datagen.flow_from_directory('/content/xray_dataset_covid19/xray_dataset_covid19/train',target_size=(224,224),batch_size=32,class_mode='binary')
```

```python
testing_set=test_datagen.flow_from_directory('/content/xray_dataset_covid19/xray_dataset_covid19/test',target_size=(224,224),batch_size=12,class_mode='binary')
```

```python
from tensorflow.keras.applications.vgg19 import VGG19
```

```python
base_model=VGG19(input_shape=(224,224,3),
                 include_top=False,
                 weights='imagenet')
```

```python
for layer in base_model.layers:
  layer.trainable=False
```

```python
from tensorflow.keras.layers import Flatten,Dense,Dropout
```

```python
x=Flatten()(base_model.output)
x=Dense(512,activation='relu')(x)
x=Dropout(0.2)(x)
x=Dense(1,activation='sigmoid')(x)
```

```python

model=tensorflow.keras.Model(base_model.input,x)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
hist=model.fit(x=training_set,steps_per_epoch=5,epochs=5)
```

```python
print(model.evaluate(x=testing_set))
```
